```python
import pandas as pd
```

```python
df1=pd.read_csv('/content/FuelConsumptionCo2.csv')
```

```python
df1.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df1.isnull().sum()
```

```python
x=df1.iloc[:,[4,5,8,9,10,11]]
```

```python
y=df1.iloc[:,-1]
```

```python
x
```

```python
from sklearn import datasets, linear_model, metrics
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,
                                                    random_state=1)
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
y_pred=reg.predict(X_test)
print(reg.intercept_)
print(reg.coef_)
print('MSE',metrics.mean_squared_error(y_test, y_pred))
print('R_squred_error',metrics.r2_score(y_test, y_pred))

 
```

```python
df3 = pd.get_dummies(df1, columns = ['FUELTYPE'])
```

```python
df3.drop(['MODELYEAR','MAKE','VEHICLECLASS','TRANSMISSION','MODEL','FUELTYPE_Z'],axis=1,inplace=True)
```

```python
df3
```

```python
Y=df3['CO2EMISSIONS'].values
X=df3.drop(['CO2EMISSIONS'],axis=1)
```

```python
from sklearn import datasets, linear_model, metrics
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=1)
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
y_pred=reg.predict(X_test)
print(reg.intercept_)
print(reg.coef_)
print('MSE',metrics.mean_squared_error(y_test, y_pred))
print('R_squred_error',metrics.r2_score(y_test, y_pred))
```
